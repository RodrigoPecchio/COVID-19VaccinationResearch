```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

covid_cases = pd.read_csv("../input/covid19-data-daily/covid19_data_final.csv")
covid_cases.head()
```

```python
covid_cases.plot.scatter(x='date',y='new_deaths', title='New COVID-19 Related Daily Deaths Globally (Jan 1, 2020 - May 16, 2021)')
```

```python
covid_cases.plot.scatter(x='date',y='new_vaccinations', title='New COVID-19 Daily Vaccinations Globally (Jan 1, 2020 - May 16, 2021)')
```

```python
covid_cases.plot.scatter(x='date',y='new_cases', title='New COVID-19 Positive Daily Cases Globally (Jan 1, 2020 - May 16, 2021)')
```

```python
covid_cases.plot.scatter(x='new_vaccinations',y='new_cases', title='COVID-19 Daily Vaccinations vs Daily New Cases Globally (Jan 1, 2020 - May 16, 2021)')
correlation = covid_cases['new_vaccinations'].corr(covid_cases['new_cases'])
print("Correlation between new daily vaccinations and new daily cases:\n" + str(correlation))
```

```python
covid_cases.plot.scatter(x='new_vaccinations',y='new_deaths', title='COVID-19 Daily Vaccinations vs Daily New COVID-19 Related Deaths Globally (Jan 1, 2020 - May 16, 2021)')
correlation = covid_cases['new_vaccinations'].corr(covid_cases['new_deaths'])
print("Correlation betwen new daily vaccinations and new daily deaths:\n" + str(correlation))
```
